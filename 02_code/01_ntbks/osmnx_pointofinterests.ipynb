```python
# Definir el lugar de búsqueda y el tipo de puntos de interés
lugar = "Medellín, Colombia"

# Se realiza un loop sobre el diccionario osm_tags para ir descargando las coordenadas de cada tipo de recurso.
for k, v in osm_tags.items():

    for amenity in v:
        
        try:

            # Definir las categorías de amenities, como "hospital", "pharmacy", "restaurant", etc.
            tags = {k: amenity}

            # Descargar los POIs desde OpenStreetMap
            pois = ox.geometries_from_place(lugar, tags)

            pois.to_file(f"puntos_interes_medellin_{k}_{amenity}.geojson", driver="GeoJSON")

        except:
            print(f"error {amenity}")
```
